```python
data_sample = df_diag['Diagnoseexport'].sample(5000,random_state = 1)
```

```python
def spacy_initial(data_sample_series):
  '''
  create a list of list of token after tokenizition in spacy
  '''
  text_sample = [text for text in data_sample_series.values ]
  word_list = {}
  word_list['text'] = []
  word_list['is_character'] = []

  for text in text_sample:
    doc = nlp(text)
    for token in doc:
      if token.is_punct is False and token.is_stop is False and token.pos_ !='SPACE':
        word_list['text'].append(token.text)
        word_list['is_character'].append(token.is_alpha)
      
  return pd.DataFrame.from_dict(word_list,orient='columns')
```

```python
word_list = spacy_initial(data_sample)
```

```python
abbrevation = word_frequency.loc[(word_frequency['is_character']==True) & (word_frequency['length of word']>1)]
abbrevation
```

```python
import re
def abb_pattern(str_):
  pattern = re.compile('^[A-Z]{2,}')
  matched = re.findall(pattern,str_)
  if matched:
    return matched[0]
  return 0
```

```python
def distribute_task(series):
  '''
  this is a funtion to seperate data into pipline to extend the capacity of spacy
  '''
  iter = len(series)//5000 +1
  df =pd.DataFrame()
  for i in range(iter) :
    if i != iter-1:
      df = pd.concat([df,spacy_initial(series[5000*i:5000*(i+1)])],axis=0)
      print('parsing...')
    else :
      df = pd.concat([df,spacy_initial(series[5000*i:])],axis=0)
  return df
```
